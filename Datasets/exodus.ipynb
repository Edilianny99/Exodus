```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
import pandas as pd
from selenium.webdriver import Chrome
```

```python
# Configurar el navegador
print("Iniciando navegador...")
driver = webdriver.Chrome(service=webdriver.chrome.service.Service(driver_path))


```

```python
try:
    # Navegar a la página de exploración
    print("Navegando a la página de Explore...")
    driver.get("https://x.com/explore")
    time.sleep(50)

    # Buscar un hashtag
    print("Buscando la barra de búsqueda...")
    search_box = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '//input[@placeholder="Buscar"]'))  # Ajusta si es necesario
    )
    print("Barra de búsqueda encontrada. Escribiendo hashtag...")
    search_box.send_keys("#SOSVenezuela")  
    search_box.send_keys(Keys.RETURN)
    print("Hashtag buscado. Esperando resultados...")
    time.sleep(50)

    # Desplazamiento para cargar más tweets
    tweets_data = []
    print("Desplazándose para cargar más tweets...")
    scroll_count = 0

    max_scroll = 15

    while len(tweets_data) < 30 and scroll_count < max_scroll:
        # Desplazar hacia abajo
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(15)  # Esperar para que carguen más tweets
        scroll_count += 1
        print(f"Desplazamiento {scroll_count} realizado...")

        # Extraer tweets visibles
        tweets = driver.find_elements(By.XPATH, '//article[@data-testid="tweet"]')
        print(f"Se encontraron {len(tweets)} tweets visibles.")

        for tweet in tweets[len(tweets_data):]:  # Solo procesa nuevos tweets
            try:
                # Extraer el bloque de contenido principal
                content_block = tweet.find_element(By.XPATH, './/div[@data-testid="tweetText"]')

                # Procesar todo el contenido dentro del bloque
                spans = content_block.find_elements(By.XPATH, './/span')
                images = content_block.find_elements(By.XPATH, './/img[@alt]')
                links = content_block.find_elements(By.XPATH, './/a')

                # Combinar texto de spans
                text = ' '.join([span.text for span in spans])
                # Combinar emojis
                emojis = ''.join([img.get_attribute('alt') for img in images])
                # Combinar hashtags y enlaces
                hashtags = ' '.join([link.text for link in links if link.text.startswith('#')])

                # Combinar todo el contenido
                full_content = f"{text} {emojis} {hashtags}".strip()

                # Extraer usuario
                username = tweet.find_element(By.XPATH, './/div[2]/div[1]//span').text
                # Extraer fecha
                date = tweet.find_element(By.XPATH, './/time').get_attribute('datetime')
                
                # Guardar los datos en la lista
                tweets_data.append({"usuario": username, "fecha": date, "contenido": full_content})
                print(f"Tweet {len(tweets_data)} extraído: {full_content} - {username} - {date}")

                # Detener si ya tienes 20 tweets
                if len(tweets_data) >= 20:
                    break
            except Exception as e:
                print(f"Error al extraer datos de un tweet: {e}")

    # Guardar los datos en un archivo CSV
    if tweets_data:
        print("Guardando tweets en un archivo CSV...")
        df = pd.DataFrame(tweets_data)
        df.to_csv("tweets_dataset4.csv", index=False, encoding='utf-8')
        print(f"Se guardaron {len(tweets_data)} tweets en tweets_dataset4.csv")
    else:
        print("No se recolectaron tweets.")

except Exception as e:
    print(f"Error general encontrado: {e}")
    input("Presiona Enter para cerrar el navegador y revisar manualmente.")

finally:
    print("Cerrando el navegador...")
    driver.quit()
```

```python
from transformers import pipeline


classifier = pipeline('sentiment-analysis', 
                      model="nlptown/bert-base-multilingual-uncased-sentiment")

sentences = ["¡Me encantan los artículos de Medium, son lo máximo!", 
                "Odio los lunes, no sirven para nada.", 
                "El libro es medio bueno; me gustaron algunos personajes."]

results = classifier(sentences)
for result in results:
    print(f"polaridad: {result['label']}, score: {round(result['score'], 4)}")
```
